## *VS Code*

### 5. Add your name to the schedule table to be a discussion leader
- Open README.md in VS Code
- Locate the row in the table for the date where you want to be a discussion leader
- Add your name to the appropriate cell (the far right, inside the last set of vertical pipes)
- Save the file so it shows up in the list of changes
- Write a descriptive commit message
- Commit your changes to your local repository
- Sync your changes with your remote repository

## *GitHub.com*

### 6. Make a pull request to add your name to the upstream repository
- Go to your remote repository; this is a downstream fork of the course repository
- Click "Sync fork" in the upper right of the central column
- Click "Contribute to the left of "Sync fork"
- Click the green "Open pull request" button
- Add a descriptive title (e.g., "Added \[your name\] to the discussion leader schedule")
- Click the green "Create pull request" button at the bottom
