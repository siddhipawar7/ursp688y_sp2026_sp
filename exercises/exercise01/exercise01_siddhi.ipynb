```python
# Import dependencies (code you need that's stored outside this file)
import json
```

```python
# Load the service call data from the json file
with open('service_requests.json', 'r') as json_file:
    service_requests = json.load(json_file)
```

```python
# Investigate the data

# What's the data type
type(service_requests)
```

```python
# How many elements are in the list?
len(service_requests)
```

```python
# What does the first element contain?
service_requests[0]
```

```python
# 1. How many times did each type of request get made?

#Main Objective: Create an empty dictionary to store the "Request Type" and the count for number of requests (key:value)

#Make empty dictionary
request_type_count = {}

#Loop through each "Request Type" in service_request
for request in service_requests:
    request_type = request["Request Type"]

#If a "Request Type" is already in the dictionary, add 1 to its existing count
    if request_type in request_type_count:
        request_type_count[request_type] += 1

#If a "Request Type" is not in the dictionary, add it to the dictionary with a value of 1
    else:
        request_type_count[request_type] = 1

#Print "Request Type" counts
print (request_type_count)


```

```python
# 2. How many times did requests get made from each city?

#Main Objective: Create an empty dictionary to store the "City" in service_requests and get the count for each city (key:value)

#Make empty dictionary
city_request_count = {}

#Loop through each "City" in service_request
for city in service_requests:
    city_request = city["City"]

#If a "City" is already in the dictionary, add 1 to its existing count
    if city_request in city_request_count:
        city_request_count[city_request] += 1

#If a "City" is not in the dictionary, add it to the dictionary with a value of 1
    else:
        city_request_count[city_request] = 1

#Print "City" counts
print (city_request_count)
```
